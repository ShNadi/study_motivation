```python
import preparation
```

```python
# Read detaset
df_2014, df_2015, df_2016 = preparation.read_ds()
```

```python
# Remove null values
preparation.dropnull(df_2014, df_2015, df_2016)

# conacatenat datasets
df = preparation.concat(df_2014, df_2015, df_2016)
```

```python
# Distribution

print(df['bsa_dummy'].value_counts())
df['bsa_dummy'].value_counts().plot.bar()
```

```python
print(df.groupby(['year']).bsa_dummy.value_counts())
df.groupby(['year','bsa_dummy']).size().unstack(fill_value=0).plot.bar()
```
